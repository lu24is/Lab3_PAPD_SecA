### ¿Cuántos Municipios tienen más de 100000 habitantes?

```python
# Filtrar las filas donde se tiene una población mayor a 100,000
Mun_mayor_100k = dataset[dataset['Poblacion'] > 100000]

# Del filtrado anterior, obtenemos el numero de municipios
num_Mun_mayor_100k = Mun_mayor_100k['Municipio'].nunique()

print("Número de municipios con población mayor a 100,000:", num_Mun_mayor_100k)
```

### Realice una gráfica de barras sobre la polbación de cada ciudad, ordenela de menor a mayo y responda: a. ¿Cuál es la segunda ciudad más poblada? b. ¿Qué posición ocupa Granada en el ranking de las ciudades más pobladas?

```python
# Ordenamos el dataset por población de manera ascendente
dataset_ascendente = dataset.sort_values(by='Poblacion')

# Creamos la gráfica de barras
plt.figure(figsize = (10,6))
plt.barh(dataset_ascendente['Municipio'], dataset_ascendente['Poblacion'])
plt.xlabel('Población')
plt.ylabel('Ciudad')
plt.title('Población de cada ciudad')
plt.grid(axis='x')
plt.show()
```

```python
# a. Obtener la segunda ciudad más poblada
segunda_ciudad = dataset_ascendente['Municipio'].iloc[-2]
print("La segunda ciudad más poblada es:", segunda_ciudad)

# b. Obtener la posición de Granada en el ranking de las ciudades más pobladas
posicion_granada = dataset_ascendente[dataset_ascendente['Municipio'] == 'Granada'].index[0] + 1
print("Granada ocupa la posición ", posicion_granada, " en el ranking de las ciudades más pobladas.")
```

### ¿Cuántos municipios de Navarra tienen más de 5000 habitantes?

### ¿Cuál es el municipio situado más al Norte? (Usar el valor de la coordenada "Y" que representa la latitud en grados). Proporcione también la provincia a la que pertenece y su población.

```python
# Encontrar el municipio situado más al Norte
municipio_mas_al_norte = dataset.loc[dataset['Y'].idxmax()]

print("Municipio:", municipio_mas_al_norte['Municipio'])
print("Provincia:", municipio_mas_al_norte['Provincia'])
print("Población:", municipio_mas_al_norte['Poblacion'])
```

### Obtenga la media, mediana, desviación estándar, valor máximo y valor mínimo de la población de los municipios de la provincia de Granada.

```python
# Filtrar los municipios de la provincia de Granada
municipios_granada = dataset[dataset['Provincia'] == 'Granada']

# Calcular lo solicitado
media = municipios_granada['Poblacion'].mean()
mediana = municipios_granada['Poblacion'].median()
desviacion_estandar = municipios_granada['Poblacion'].std()
valor_maximo = municipios_granada['Poblacion'].max()
valor_minimo = municipios_granada['Poblacion'].min()

print("Media:", media)
print("Mediana:", mediana)
print("Desviación estándar:", desviacion_estandar)
print("Valor máximo:", valor_maximo)
print("Valor mínimo:", valor_minimo)
```

### Realice un histograma con la población de los Municipios para cada una de las provincias.

```python
# Calcular el histograma de población para cada provincia
plt.figure(figsize=(10, 6))
for provincia in dataset['Provincia'].unique():
    poblacion_provincia = dataset[dataset['Provincia'] == provincia]['Poblacion']
    plt.hist(poblacion_provincia, bins=20, alpha=0.5, label=provincia)
    plt.title('Histograma de Población por Provincia')
    plt.xlabel('Población')
    plt.ylabel('Frecuencia')
    plt.legend(title='Provincia')
    plt.show()
```

### Seleccione al azar cincuenta municipios diferentes de entre los diponibles en el archivo. Asegúrese de que no se repitan. Luego calcule el promedio de la población y la desviación estándar de esto 50 municipios

```python
municipios_aleatorios = dataset.sample(n=50, replace=False)

# Calcular el promedio y la desviación estándar de la población de los 50 municipios seleccionados
promedio_poblacion = municipios_aleatorios['Poblacion'].mean()
desviacion_estandar_poblacion = municipios_aleatorios['Poblacion'].std()

# Mostrar los resultados
print("Promedio de la población de los 50 municipios seleccionados:", promedio_poblacion)
print("Desviación estándar de la población de los 50 municipios seleccionados:", desviacion_estandar_poblacion)
```

```python
from geopy.distance import geodesic

# Introduzco las coordenadas de Madrid
coord_madrid = (40.416630, -3.703797)

# Calcular la distancia entre Madrid y cada municipio, excluyendo Madrid
distancias = {}
for index, row in dataset.iterrows():
    if row['Municipio'] != 'Madrid':  # Excluir Madrid del cálculo
        coord_municipio = (row['Y'], row['X'])
        distancia = geodesic(coord_madrid, coord_municipio).kilometers
        distancias[row['Municipio']] = distancia

# Encontrar el municipio más cercano y el más lejano
municipio_mas_cercano = min(distancias, key=distancias.get)
municipio_mas_lejano = max(distancias, key=distancias.get)

# Mostrar los resultados
print("Municipio más cercano a Madrid:", municipio_mas_cercano)
print("Distancia al municipio más cercano:", distancias[municipio_mas_cercano], "kilómetros")
print("Municipio más lejano a Madrid:", municipio_mas_lejano)
print("Distancia al municipio más lejano:", distancias[municipio_mas_lejano], "kilómetros")
```
